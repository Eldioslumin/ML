### Pivot tables and graphics

We want to see how the instances are distributed between two categories: 'International plan' and 'Churn'. 

<b>Exercise 7:</b>
* Build the cross table between the features using the method <i>crosstab</i>.
* Visualize the distribution for the feature 'Churn', depending on the value of the features 'International plan', 'Voice mail plan', and 'Customer service calls'.
<b>Hint:</b> Use commands plt.sublot and sns.subplots.

Your plot should look something like this:
<img src="dist_crosstab.png">

What do you see? What conclusions can be drawn? What feature (intuitively) can be more important for Churn prediction?

```python
### Write Your code here ###
cross1 = pd.crosstab(telecom_churn['Churn'], telecom_churn["International plan"],)
cross2 = pd.crosstab(telecom_churn['Churn'], telecom_churn["Voice mail plan"],)
cross3 = pd.crosstab(telecom_churn['Churn'], telecom_churn["Customer service calls"])

fig, axs = plt.subplots(1, 3, figsize=(20, 8))

cross1.plot(kind='bar', ax=axs[0])
axs[0].set_xlabel("International Plan")
axs[0].set_ylabel("Number of Clients")

cross2.plot(kind='bar', ax=axs[1])
axs[1].set_xlabel("Voice Mail Plan")
axs[1].set_ylabel("Number of Clients")

cross3.plot(kind='bar', ax=axs[2])
axs[2].set_xlabel("Service Calls")
axs[2].set_ylabel("Number of Clients")

plt.tight_layout()
plt.show()
```
